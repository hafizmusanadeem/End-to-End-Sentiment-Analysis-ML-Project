**This dataset is downloaded from kaggle, It included 4 text files and 1 csv file. Firstly each text file is converted into csv resulting in total 5 csv files.**
**There are total 3 sentiments to predict -> [positive , negative, neutral]**
**Each csv file had text with unequal number of sentiments**
**Therefore, a dataset was needed with texts possessing equal number of sentiments**

**Result:**
**A Dataset with 3000 rows and 2 columns (text, sentiment), where each sentiment[positive, negative, neutral] is present 1000 times**

# **Data Collection**

```python
import re
import pandas as pd

class DataIngestion:

    def openfile(self, path):
        with open(file=path, mode="r", encoding="latin-1",) as file:
            data = file.read()
            # print(data)
        return data


    def traverse(self, text):
        """
        Split the text at occurrences of 'positive', 'neutral', or 'negative'.
        """
        parts = re.split(r'(positive|neutral|negative)', text)
        parts = [p.strip() for p in parts if p.strip()]
        return parts
    
    def new_data(self, text):
        if len(text) % 2 != 0:
            text = text[:-1]

        texts = text[0::2]
        sentiment = text[1::2]

        # Remove '@', '\n' from sentiments
        texts = [t.replace("\n", " ").replace(".@","").strip() for t in texts]

        data = pd.DataFrame({
            "texts":texts,
            "sentiment":sentiment
        })

        return data
```

# **Creating new data**

```python
version_1 = pd.read_csv("version_1.csv")
version_2 = pd.read_csv("version_2.csv")
version_3 = pd.read_csv("version_3.csv")
version_4 = pd.read_csv("version_4.csv")
```

```python
versions = [version_1, version_2, version_3, version_4]
```

```python
df = pd.concat(versions, ignore_index = True)
```

```python
df.shape
```

```python
sentiment_classes = df['sentiment'].unique()
```

```python
final_df = pd.concat([
    df[df['sentiment'] == cls].sample(n = 1000, random_state=42) for cls in sentiment_classes
])
```

```python
final_df.shape
```

```python
final_df = final_df.sample(frac = 1, random_state=42).reset_index(drop=True)
```

```python
final_df.to_csv('final_balanced_dataset.csv', index = False)
print("Balanced dataset created successfully!")
```
